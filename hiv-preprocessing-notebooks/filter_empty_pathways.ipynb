```python
import os
```

```python
hiv_preprocessing_results_path = "C:/Users/sumed/BDS-SROP-24/hiv-benchmarking-05-060-modified"
```

```python
def delete_empty_pathway_files_and_dirs(root_dir):
    no_of_files_deleted = 0
    for dirpath, dirnames, filenames in os.walk(root_dir, topdown=False):
        for filename in filenames:
            # generate file path
            file_path = os.path.join(dirpath, filename)
            # check if file exists and is empty (size = 0)
            if os.path.isfile(file_path) and os.path.getsize(file_path) == 0:
                # print the path of the file that will be deleted
                print("Deleting File >>>", file_path.replace('\\', '/'))
                # delete empty file
                os.remove(file_path)
                no_of_files_deleted += 1
        # check if directory is empty after deleting files
        if not os.listdir(dirpath):
            # print the path of the directory that will be deleted
            print("Deleting Directory >>>", dirpath.replace('\\', '/'))
            # delete empty directory
            os.rmdir(dirpath)
    print(no_of_files_deleted, "file(s) have been deleted.")
```

```python
delete_empty_pathway_files_and_dirs(hiv_preprocessing_results_path)
```
